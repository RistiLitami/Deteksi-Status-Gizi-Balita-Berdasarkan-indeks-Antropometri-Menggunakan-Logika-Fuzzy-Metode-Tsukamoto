## Implementasi Program Python

Berikut merupakan baris kode yang ditujukan untuk user, agar user bisa memasukkan data-data dari balita yang diperlukan dan akan dikelolah oleh program hingga mendapatkan hasil berupa Status Gizi daripada balita tersebut.

```python
# Input Program

nama_balita = input("Masukkan Nama Balita : ")
usia_balita = input("Masukkan Usia Balita (tahun) : ")
jenis_kelamin = input("Masukkan Jenis Kelamin Balita (L/P) : ")
tb_balita = input("Masukkan Tinggi Badan Balita : ")
bb_balita = input("Masukkan Berat Badan Balita : ")
```

Baris kode berikut merupakan inisialisasi dan penamaan pada variabel-variabel yang akan digunakan di dalam program.

```python
# inisialisasi
value_pendek = float()
value_normal = float()
value_tinggi = float()

value_kurang = float()
value_bbnormal = float()
value_lebih = float()

tb = int(tb_balita)
bb = int(bb_balita)
usia = int(usia_balita)
```

Baris kode berikut ini merupakan proses fuzifikasi dari input 'Tinggi Badan'. Proses Fuzifikasi ini didapatkan dari rumus Plotting Fungsi Keanggotaan 'Tinggi Badan'. Data Tinggi Badan dari balita yang sudah diinputkan oleh user akan masuk ke tahap proses fuzifikasi pada bagian kode ini. Dapat dilihat pada kode berikut ini, terdapat kondisi if dengan batas-batas nilai yang sudah ditentukan sebelumnya. Setelah diproses maka, kode berikut akan mengeluarkan output berupa fungsi keanggotaan dari nilai Tinggi Badan yang diinputkan oleh user.

```python
# Proses Fuzifikasi Tinggi Badan
if tb < 45 :
    value_pendek = 1
    value_normal = 0
    value_tinggi = 0
if tb > 45 and tb < 75 :
    value_pendek = (75-tb)/(75-45)
    value_normal = (tb-45)/(75-45)
    value_tinggi = 0
if tb == 75 :
    value_pendek = 0
    value_normal = 1
    value_tinggi = 0
if tb > 75 and tb < 101 :
    value_pendek = 0
    value_normal = (101-tb)/(101-75)
    value_tinggi = (tb-75)/(101-75)
if tb > 101 :
    value_pendek = 0
    value_normal = 0
    value_tinggi = 1
print(f"Maka Tinggi Badan Laki- Laki dalam variabel linguistik, derajat keanggotaan adalah")
print(f'Pendek : ', value_pendek)
print(f'Normal : ', value_normal)
print(f'Tinggi', value_tinggi)
```

Sama halnya dengan baris kode sebelumnya, baris kode berikut merupakan proses fuzifikasi dari 'Berat Badan'. Baris kode ini juga dibuat berdasarkan hasil plotting membership function daripada 'Tinggi Badan'. Prosesnya juga sama seperti baris kode sebelumnya, inputan Berat Badan dari user akan dikelolah di baris kode ini hingga menghasilkan output berupa nilai membership function dari Berat Badan yang telah diinputkan.

```python
# Proses Fuzifikasi Berat Badan
if bb < 7 :
    value_kurang = 1
    value_bbnormal = 0
    value_lebih = 0
if bb > 7 and bb < 13 :
    value_kurang = (13-bb)/(13-7)
    value_bbnormal = (bb-7)/(13-7)
    value_lebih = 0
if bb == 13 :
    value_kurang = 0
    value_bbnormal = 1
    value_lebih = 0
if bb > 13 and bb < 19 :
    value_kurang = 0
    value_bbnormal = (19-bb)/(19-13)
    value_lebih = (bb-13)/(19-13)
if bb > 19 :
    value_kurang = 0
    value_bbnormal = 0
    value_lebih = 1
print("Maka Berat Badan dalam variabel linguistik, derajat keanggotaan adalah")
print('Pendek : ', value_kurang)
print('Normal : ', value_bbnormal)
print('Tinggi', value_lebih)
```

Baris kode berikut merupakan, proses Inferensi yang dimana merupakan proses dengan menggunakan fungsi implikasi MIN untuk mendapatkan nilai a-predikat tiap-tiap  rule  $(a1,a2,  a3,  ...  an)$. Kemudian masing-masing nilai a-predikat ini digunakan untuk menghitung keluaran hasil inferensi secara tegas $(crisp)$ masing-masing rule $(z1, z2, z3, ...zn)$.

```python
# Proses Inferensi Status Gizi
status_gizi=[]
def giziburuk (variabel_tb, variabel_bb):
    if variabel_tb != 0:
        if variabel_bb != 0:
            hasil_output = min(variabel_tb, variabel_bb)
            status_gizi.append([hasil_output,43])

def gizikurang (variabel_tb, variabel_bb):
    if variabel_tb != 0:
        if variabel_bb != 0:
            hasil_output = min(variabel_tb, variabel_bb)
            status_gizi.append([hasil_output,48])

def gizinormal (variabel_tb, variabel_bb):
    if variabel_tb != 0:
        if variabel_bb != 0:
            hasil_output = min(variabel_tb, variabel_bb)
            status_gizi.append([hasil_output,53])

def gizilebih (variabel_tb, variabel_bb):
    if variabel_tb != 0:
        if variabel_bb != 0:
            hasil_output = min(variabel_tb, variabel_bb)
            status_gizi.append([hasil_output,70])

def obesitas (variabel_tb, variabel_bb):
    if variabel_tb != 0:
        if variabel_bb != 0:
            hasil_output = min(variabel_tb, variabel_bb)
            status_gizi.append([hasil_output,83])
```

   Baris kode berikut merupakan proses pembentukan basis pengetahuan Fuzzy  **(Rule dalam bentuk IF...THEN)**, yaitu $Ʃ$ Secara umum bentuk model Fuzzy  Tsukamoto  adalah  **IF  (X  IS  A) and (Y IS B) and (Z IS C)** , dimana A,B, dan C adalah himpunan fuzzy.
    
   Disini, karena balita tersebut memiliki 5 tahapan usia dengan rentang usia yang sudah ditentukan, maka saya membuat fungsi def pertahap dengan aturan fuzzy yang sudah ditetapkan pada tabel di dokumen

```python
# Proses Aturan Fuzzy berdasarkan tahapan usia
def tahap1():
    gizinormal(value_kurang, value_pendek)
    gizinormal(value_kurang, value_normal)
    gizikurang(value_kurang, value_tinggi)
    gizilebih(value_bbnormal, value_pendek)
    gizilebih(value_bbnormal, value_normal)
    gizilebih(value_bbnormal, value_tinggi)
    gizilebih(value_lebih, value_pendek)
    gizilebih(value_lebih, value_normal)
    obesitas(value_lebih, value_tinggi)
    print("Maka, Status Gizi Balita adalah ", status_gizi)

def tahap2():
    gizikurang(value_kurang, value_pendek)
    gizikurang(value_kurang, value_normal)
    gizikurang(value_kurang, value_tinggi)
    gizinormal(value_bbnormal, value_pendek)
    gizinormal(value_bbnormal, value_normal)
    gizinormal(value_bbnormal, value_tinggi)
    gizilebih(value_lebih, value_pendek)
    gizilebih(value_lebih, value_normal)
    obesitas(value_lebih, value_tinggi)
    print("Maka, Status Gizi Balita adalah ", status_gizi)

def tahap3():
    giziburuk(value_kurang, value_pendek)
    giziburuk(value_kurang, value_normal)
    gizinormal(value_kurang, value_tinggi)
    gizinormal(value_bbnormal, value_pendek)
    gizinormal(value_bbnormal, value_normal)
    gizinormal(value_bbnormal, value_tinggi)
    gizilebih(value_lebih, value_pendek)
    gizilebih(value_lebih, value_normal)
    obesitas(value_lebih, value_tinggi)
    print("Maka, Status Gizi Balita adalah ", status_gizi)

def tahap4():
    gizikurang(value_kurang, value_pendek)
    gizikurang(value_kurang, value_normal)
    gizikurang(value_kurang, value_tinggi)
    gizinormal(value_bbnormal, value_pendek)
    gizinormal(value_bbnormal, value_normal)
    gizinormal(value_bbnormal, value_tinggi)
    gizilebih(value_lebih, value_pendek)
    gizilebih(value_lebih, value_normal)
    gizinormal(value_lebih, value_tinggi)
    print("Maka, Status Gizi Balita adalah ", status_gizi)

def tahap5():
    giziburuk(value_kurang, value_pendek)
    giziburuk(value_kurang, value_normal)
    giziburuk(value_kurang, value_tinggi)
    gizikurang(value_bbnormal, value_pendek)
    gizikurang(value_bbnormal, value_normal)
    gizikurang(value_bbnormal, value_tinggi)
    gizilebih(value_lebih, value_pendek)
    gizilebih(value_lebih, value_normal)
    gizinormal(value_lebih, value_tinggi)
    print("Maka, Status Gizi Balita adalah ", status_gizi)
```

Baris kode berikut merupakan kondisi yang akan dijalankan sesuai dengan rentang nilai usia daripada balita. Rentang usia pada balita dapat dilihat pada tabel sebelumnya.

```python
# Proses inferensi yang akan dijalankan berdasarkan range tahap usia
if usia >= 0 and usia <= 1 :
    tahap1()
if usia > 1 and usia <= 2 :
    tahap2()
if usia > 2 and usia <= 3 :
    tahap3()
if usia > 3 and usia <= 4 :
    tahap4()
if usia > 4 and usia <= 5 :
    tahap5()
```

Baris kode berikut merupakan proses Defuzzifikasi, yang mana merupakan langkah terakhir dalam suatu sistem logika fuzzy dengan tujuannya mengkonversi setiap hasil dari inference engine yang diekspresikan dalam bentuk fuzzy set kesuatu bilangan real. Hasil konversi tersebut merupakan aksi yang diambil oleh sistem kendali logika fuzzy.

Karena project ini menggunakan metode Tsukamoto maka, proses Defuzzyfikasinya menggunakan metode rata-rata (Average) : 
$Z=(Ʃa1.Z1)/Ʃa1$

Dari proses defuzifikasi ini, output yang berupa nilai status gizi dari data-data balita akan tampil. Ouput yang keluar di baris kode ini berupa angka.

```python
# Defuzifikasi
perkalian = 0
pembagian = 0

for j in range (0, len(status_gizi)):
    kali = status_gizi[j][0]*status_gizi[j][1]
    bagi = status_gizi[j][0]
    perkalian = perkalian + kali
    pembagian = pembagian + bagi
z = perkalian / pembagian
print("Maka nilai status gizi adalah", z)
```

Berikut ini merupakan output atau keluaran yang berupa huruf. Sebagaimana nilai z dari proses defuzifikasi sebelumnya merupakan angka, di baris kode ini dilakukan penerjemahan angka ke dalam huruf agar mudah untuk dipahami oleh user. Di baris kode ini digunakan pengkondisian IF dengan nilai range berdasarkan domain dari Status Gizi.

```python
# Output untuk status Gizi
if z <= 48 :
    print("Status Gizi Balita : Buruk")
if z > 48 and z <= 53 :
    print("Status Gizi Balita : Kurang")
if z > 53 and z <= 70 :
    print("Status Gizi Balita : Normal")
if z > 70 and z <= 83 :
    print("Status Gizi Balita : Lebih")
if z > 83 and z <= 123 :
    print("Status Gizi Balita : Obesitas")
```
